### leetcode 35

주어진 정렬된 리스트가 있을 때 `target 값이 삽입될 인덱스`를 반환하는 문제.

반을 어떻게 나눌 것인가, 인덱스를 어떻게 최신화 시켜줄 것인가 를 생각해야하는 문제.


### 코드

```python
class Solution(object):
    def searchInsert(self, nums, target):
        if target <= nums[0]:
            return 0
        
        elif target > nums[-1]:
            return len(nums)
        
        left = 0
        right = len(nums) - 1
        
        while left != right - 1:
            mid = (left + right) // 2
            if target > nums[mid]:
                left = mid
            elif target < nums[mid]:
                right = mid
            else:
                return mid
        
        return right
```

### 이해하기

nums = [1,3,5,6], target = 2 라고 했을 때 하나씩 살펴보면

2 가 첫 번째 값보다 작으면 0 번 인덱스에 추가되어야 하기 때문에 인덱스 0 을 리턴

2 가 마지막 값보다 크면 len(nums) 으로 target 값이 추가될 인덱스 nums의 길이를 리턴

위의 두 경우가 아니라면 반 씩 잘라가며 찾아야 하는데,

인덱스의 시작과 끝을 포인터로 잡고 mid 값을 최신화 해나가야 함.

left = 0, right = 3

첫 번째 루프에서

3 // 2 몫 1 = mid

2 < nums[1] = 3 이고, right = 1

left = 0, right = 1 

두 번째 루프에서 

left == right - 1 이기 때문에 

return right = 1 함.
